```python
# Generate waveform
T_obs = np.pi * 1e5
T_obs_year = T_obs/(365*24*60*60)  # Convert seconds to years
dt = 5

sim_t = np.arange(0, T_obs, dt)

A = 1e-22
f_0 = 1e-3
dotf_0 = 1e-10


true_params = [A, f_0, dotf_0] 
h = waveform(true_params, sim_t)
# window_check = scipy.signal.windows.tukey(len(h), alpha=0.1)
h_zp = zero_pad(h)

N_t_full = len(h_zp)

h_fft  = np.fft.rfft(h_zp)
```

```python
variance_noise_f = (N_t_full * psd_A/(4*dt))

def noise_gen(seed = 1, mask = None):
    np.random.seed(seed)
    noise_f = (np.random.normal(0, np.sqrt(variance_noise_f)) +
                1j * np.random.normal(0, np.sqrt(variance_noise_f)))

    variance_noise_f_DC = (N_t_full * psd_A[0]/(2*dt)) 
    variance_noise_f_nyq = (N_t_full * psd_A[-1]/(2*dt)) 

    noise_f_DC = np.random.normal(0,np.sqrt(variance_noise_f_DC), 1)
    noise_f_nyq = np.random.normal(0,np.sqrt(variance_noise_f_nyq), 1)

    noise_f[0] = noise_f_DC
    noise_f[-1] = noise_f_nyq

    if mask is None:
        return noise_f
    else:
        noise_t = np.fft.irfft(noise_f)
        noise_t_masked = noise_t * mask
        noise_f_masked = np.fft.rfft(noise_t_masked)
        return noise_f_masked
```

```python
# I want to build a function that segments noise_t_mask into a list of segments
# Segment waveform into valid data sections (excluding gaps with NaNs)

def segment_data_by_gaps(signal, mask):
    """
    Segment a signal into contiguous valid data sections based on a gap mask.
    
    Args:
        signal: Input signal array
        mask: Gap mask (1 for valid data, NaN for gaps)
    
    Returns:
        segments: List of valid data segments
        segment_indices: List of (start_idx, end_idx) for each segment
        segment_times: List of time arrays for each segment (if using sim_t)
    """
    segments = []
    segment_indices = []
    segment_times = []
    
    # Convert mask to binary (1 for valid, 0 for NaN/gaps)
    valid_mask = ~np.isnan(mask)
    
    # Find transitions from gap to valid data and vice versa
    mask_diff = np.diff(np.concatenate(([0], valid_mask.astype(int), [0])))
    
    # Find start and end indices of valid segments
    starts = np.where(mask_diff == 1)[0]  # Transition from gap to valid
    ends = np.where(mask_diff == -1)[0]   # Transition from valid to gap
    
    print(f"Found {len(starts)} valid data segments:")
    
    for i, (start, end) in enumerate(zip(starts, ends)):
        segment_data = signal[start:end]
        segment_time = sim_t[start:end]
        
        segments.append(segment_data)
        segment_indices.append((start, end))
        segment_times.append(segment_time)
        
        duration_days = (segment_time[-1] - segment_time[0]) / (24 * 60 * 60)
        print(f"  Segment {i+1}: samples {start} to {end-1} ({len(segment_data)} points, {duration_days:.2f} days)")
    
    return segments, segment_indices, segment_times

noise_t_mask = np.fft.irfft(noise_f_mask)
segments, segment_indices, segment_times = segment_data_by_gaps(noise_t, mask = full_mask)

```

```python
def noise_gen_seg(variance_seg_f, seed = 1, mask = None):
    np.random.seed(seed)
    noise_f = (np.random.normal(0, np.sqrt(variance_seg_f)) +
                1j * np.random.normal(0, np.sqrt(variance_seg_f)))


    variance_noise_f_DC = 2 * variance_seg_f[0]
    variance_noise_f_nyq = 2 * variance_seg_f[-1]

    noise_f_DC = np.random.normal(0,np.sqrt(variance_noise_f_DC), 1)
    noise_f_nyq = np.random.normal(0,np.sqrt(variance_noise_f_nyq), 1)

    noise_f[0] = noise_f_DC
    noise_f[-1] = noise_f_nyq

    if mask is None:
        return noise_f
    else:
        noise_t = np.fft.irfft(noise_f)
        noise_t_masked = noise_t * mask
        noise_f_masked = np.fft.rfft(noise_t_masked)
        return noise_f_masked
```

```python
# Fill in missing data between segments using the generated noise segments
def fill_missing_data_between_segments(original_noise, segment_indices, missing_data_segments):
    """
    Fill in missing data between segments using generated noise segments.
    
    Args:
        original_noise: Original noise time series
        segment_indices: List of (start_idx, end_idx) for each valid segment
        missing_data_segments: List of noise segments to fill gaps
    
    Returns:
        filled_noise: Complete noise time series with gaps filled
    """
    filled_noise = original_noise.copy()
    
    # Fill gaps between segments
    for i in range(len(segment_indices) - 1):
        # Get end of current segment and start of next segment
        current_end = segment_indices[i][1]
        next_start = segment_indices[i+1][0]
        gap_length = next_start - current_end
        
        if gap_length > 0:  # There's a gap to fill
            # Use the corresponding missing data segment
            if i < len(missing_data_segments):
                gap_data = missing_data_segments[i]
                
                # Truncate or pad the gap data to match the exact gap length
                if len(gap_data) >= gap_length:
                    filled_noise[current_end:next_start] = gap_data[:gap_length]
                else:
                    # If gap data is shorter, repeat it to fill the gap
                    repeats = gap_length // len(gap_data) + 1
                    extended_gap_data = np.tile(gap_data, repeats)
                    filled_noise[current_end:next_start] = extended_gap_data[:gap_length]
                
                print(f"Filled gap {i+1}: indices {current_end} to {next_start-1} ({gap_length} points)")
    
    # Fill any gap before the first segment
    if len(segment_indices) > 0 and segment_indices[0][0] > 0:
        gap_length = segment_indices[0][0]
        if len(missing_data_segments) > 0:
            gap_data = missing_data_segments[-1]  # Use last segment for pre-gap
            if len(gap_data) >= gap_length:
                filled_noise[:gap_length] = gap_data[:gap_length]
            else:
                repeats = gap_length // len(gap_data) + 1
                extended_gap_data = np.tile(gap_data, repeats)
                filled_noise[:gap_length] = extended_gap_data[:gap_length]
        print(f"Filled pre-gap: indices 0 to {gap_length-1} ({gap_length} points)")
    
    # Fill any gap after the last segment
    if len(segment_indices) > 0 and segment_indices[-1][1] < len(original_noise):
        gap_start = segment_indices[-1][1]
        gap_length = len(original_noise) - gap_start
        if len(missing_data_segments) > 0:
            gap_data = missing_data_segments[-1]  # Use last segment for post-gap
            if len(gap_data) >= gap_length:
                filled_noise[gap_start:] = gap_data[:gap_length]
            else:
                repeats = gap_length // len(gap_data) + 1
                extended_gap_data = np.tile(gap_data, repeats)
                filled_noise[gap_start:] = extended_gap_data[:gap_length]
        print(f"Filled post-gap: indices {gap_start} to {len(original_noise)-1} ({gap_length} points)")
    
    return filled_noise

# Fill the missing data
filled_noise_t = fill_missing_data_between_segments(noise_t, segment_indices, missing_data_segments_t)

print(f"\nOriginal noise length: {len(noise_t)}")
print(f"Filled noise length: {len(filled_noise_t)}")
print(f"Lengths match: {len(noise_t) == len(filled_noise_t)}")
```
